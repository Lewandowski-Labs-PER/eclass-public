```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
cis = pd.read_csv('anon_cis.csv')
pre = pd.read_csv('anon_pre.csv')
post = pd.read_csv('anon_post.csv')
```

```python
first_year = cis[cis.Q18=='First year (introductory) lab'][['pre_survey_id', 'post_survey_id']]
buffy = cis[cis.Q18=='Beyond the first year lab'][['pre_survey_id', 'post_survey_id']]
```

```python
pre_qs = ['q01a', 'q01b', 'q27a', 'q27b', 'q03a', 'q03b', 'q22a', 'q22b', 'q06a', 'q06b', 'q19a', 'q19b', 'q24a', 'q24b', 'q10a', 'q10b', 'q28a', 'q28b', 'q05a', 'q05b', 'q11a', 'q11b', 'q02a', 'q02b', 'q26a', 'q26b', 'q13a', 'q13b', 'q15a', 'q15b', 'q30a', 'q30b', 'q16a', 'q16b', 'q18a', 'q18b', 'q20a', 'q20b', 'q23a', 'q23b', 'q09a', 'q09b', 'q21a', 'q21b', 'q40a', 'q40b', 'q12a', 'q12b', 'q25a', 'q25b', 'q17a', 'q17b', 'q04a', 'q04b', 'q14a', 'q14b', 'q29a', 'q29b', 'q07a', 'q07b', 'q31a', 'q31b']
```

```python
firstyear_prescores = pre[pre.survey_id.isin(first_year.pre_survey_id)][pre_qs].sum(axis=1)
buffy_prescores = pre[pre.survey_id.isin(buffy.pre_survey_id)][pre_qs].sum(axis=1)
```

```python
fig, ax = plt.subplots(figsize=(9, 6))

def get_hist(data):
    h, e = np.histogram(data, bins=np.linspace(150, 350, 100), density=True)
    return h, e

first_h, first_e = get_hist(firstyear_prescores)
ax.plot(first_e[:-1], first_h, label='First Year Lab')

buffy_h, buffy_e = get_hist(buffy_prescores)
ax.plot(buffy_e[:-1], buffy_h, label='Beyond First Year')
ax.legend(fontsize=15)
ax.set_ylabel('PDF', fontsize=15)
ax.set_xlabel('Cumulative Pre-Score', fontsize=15)
```
